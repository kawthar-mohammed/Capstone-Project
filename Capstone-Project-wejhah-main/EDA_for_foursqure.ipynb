```python
import arabic_reshaper
from bidi.algorithm import get_display

def fix_arabic_text(text):
    """
    Fixes Arabic text for proper RTL display.
    
    If the text appears to be already reshaped (i.e. it contains characters
    in the presentation forms range U+FE70 to U+FEFF), it skips reshaping 
    and only applies get_display to fix the order. Otherwise, it reshapes first.
    """
    if not isinstance(text, str):
        text = str(text)
    # Check if text already contains presentation forms
    if any('\uFE70' <= char <= '\uFEFF' for char in text):
        # Already reshaped, so only adjust the display order
        fixed = get_display(text)
    else:
        # Not yet reshaped; reshape then adjust order
        reshaped = arabic_reshaper.reshape(text)
        fixed = get_display(reshaped)
    return fixed

def fix_arabic_column(df, column_name):
    """Applies the Arabic text correction to all values in a given column."""
    df[column_name] = df[column_name].astype(str).apply(fix_arabic_text)
    return df
```

```python
from math import radians, cos, sin, sqrt, atan2
import pandas as pd
# Complete list of neighborhoods (name, lat, lon)
riyadh = [
    ("Al Sahafah", 24.803329, 46.639133), # 
    ("Al Rabi", 24.795393, 46.658502), # 
    ("Al Nada", 24.805883, 46.683010), # 
    ("Al Narjis", 24.8868, 46.6453), # 
    ("Al ‘Arid", 24.9410, 46.5580), # 
    ("Al Nafil", 24.782061, 46.673138), # 
    ("Al ‘Aqiq", 24.773696, 46.630278), # 
    ("Al Wadi", 24.789435, 46.691111), # 
    ("Al Ghadir", 24.774338, 46.655725), # 
    ("Al Yasmin", 24.825492, 46.645506), # 
    ("Al Falah", 24.796854, 46.708911), # 
    ("Banban", 25.1000, 46.5000), # 
    ("Al Qirawan", 24.835707, 46.588102), # 
    ("Hittin", 24.763594, 46.603193), # 
    ("Al Malqa", 24.800978, 46.599884), # 
    ("Al Rawdah", 24.735197, 46.766606), # 
    ("Al Rimal", 24.868298, 46.819176), # 
    ("Al Munsiyah", 24.830625, 46.768180), # 
    ("Qurtubah", 24.816610, 46.732107), # 
    ("Al Janadriyah", 24.874620, 46.914538), # 
    ("Al Qadisiyah", 24.820355, 46.832171), # 
    ("Al Yarmuk", 24.806966, 46.780283), # 
    ("Ghirnatah (Granada)", 24.792252, 46.743551), # 
    ("Ishbiliyah", 24.795295, 46.797018), # 
    ("Al Hamra", 24.777299, 46.755940), # 
    ("Al Mu'aysim (Al Mu’aizilah)", 24.794113, 46.847337), # 
    ("Al Khaleej", 24.774367, 46.802162), # 
    ("King Faisal", 24.762708, 46.775200), # 
    ("Al Quds", 24.757108, 46.753978), # 
    ("An Nahdah", 24.760764, 46.813228), # 
    ("Al Andalus", 24.743469, 46.790225), # 
    ("East Al Naseem", 24.741012, 46.842088), # 
    ("West Al Naseem", 24.725588, 46.826113), # 
    ("As Salam", 24.707980, 46.812565), # 
    ("Ar Rayyan", 24.711461, 46.777113), # 
    ("Umm Al Hammam (East)", 24.684976, 46.658778), # 
    ("Umm Al Hammam (West)", 24.690595, 46.642350), # 
    ("Al Aziziyah", 24.588095, 46.777443), # 
    ("Taybah (Al Taibah)", 24.554506, 46.814119), # 
    ("Ad Dar Al Bayda", 24.559590, 46.794021), # 
    ("Al Shifa", 24.564974, 46.699419), # 
    ("Badr", 24.526356, 46.692448), # 
    ("Al Marwah", 24.543875, 46.678347), # 
    ("Al Hazm", 24.541416, 46.649821), # 
    ("Al Mansuriyah", 24.515338, 46.790894), # 
    ("Dirab", 24.506087, 46.621640), # 
    ("Al Uraija", 24.627045, 46.658139), # 
    ("Hijrat Wadi Laban", 24.608211, 46.548163), # 
    ("Dhahrat Laban", 24.634237, 46.548403), # 
    ("Shubra", 24.574487, 46.665164), # 
    ("As Suwaidi", 24.589750, 46.669655), # 
    ("As Suwaidi Al Gharbi", 24.573917, 46.622862), # 
    ("Dhahrat Al Badiah", 24.597627, 46.645941), # 
    ("Al Badiah",24.617597, 46.677971), # 
    ("Sultana", 24.602795, 46.688851), # 
    ("Az Zahra", 24.686516, 46.730357), # 
    ("Namar", 24.576014, 46.689328), # 
    ("Dhahrat Namar", 24.559673, 46.608500), # 
    ("Tuwaiq", 24.577483, 46.556666), # 
    ("Al Malaz", 24.662771, 46.732030), # 
    ("Al Rabwah", 24.691833, 46.758159), # 
    ("Jarir", 24.677952, 46.751709), # 
    ("As Safa", 24.666869, 46.763167), # 
    ("Adh Dhubbat", 24.679357, 46.721530), # 
    ("Al Wizarat", 24.674183, 46.712200), # 
    ("Al Faruq", 24.651466, 46.771769), # 
    ("Al Amal", 24.643030, 46.719084), # 
    ("Thulaim",24.640868, 46.727866), # 
    ("Al Murabba", 24.662535, 46.707007), # 
    ("Al Futah",24.641475, 46.708730), # 
    ("Al Morooj",24.757602, 46.663114), # 
    ("Al Mursalat", 24.746842, 46.690077), # 
    ("An Nuzhah", 24.755813, 46.707651), # 
    ("Al Mughrizat", 24.763837, 46.725299), # 
    ("Al Wurud", 24.724934, 46.679873), # 
    ("King Salman Neighborhood", 24.737337, 46.705895), # 
    ("Al Olaya", 24.695890, 46.680222), # 
    ("As Sulaymaniyah", 24.703162, 46.708742), # 
    ("King Abdul Aziz District", 24.715223, 46.734389) # 
]

def haversine(lat1, lon1, lat2, lon2):
    """Calculate the great-circle distance between two points on the Earth using the Haversine formula."""
    R = 6371  # Radius of Earth in km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

def assign_nearest_neighborhood(df, neighborhoods):
    """Assigns the nearest neighborhood to each location in df based on latitude and longitude."""
    assigned_neighborhoods = []
    
    for _, row in df.iterrows():
        min_distance = float("inf")
        nearest_neighborhood = None
        
        for neighborhood, lat, lon in neighborhoods:
            distance = haversine(row['Latitude'], row['Longitude'], lat, lon)
            if distance < min_distance:
                min_distance = distance
                nearest_neighborhood = neighborhood
        
        assigned_neighborhoods.append(nearest_neighborhood)
    
    df['Nearest Neighborhood'] = assigned_neighborhoods
    return df

# Example usage:
df_coffee = assign_nearest_neighborhood(df_coffee, riyadh)
df_rest = assign_nearest_neighborhood(df_rest, riyadh)
df_venue = assign_nearest_neighborhood(df_venue, riyadh)
```

```python
import pandas as pd
import numpy as np
from math import radians, cos, sin, asin, sqrt
import ipywidgets as widgets
from IPython.display import display, clear_output

def haversine_distance(lat1, lon1, lat2, lon2):
    """Calculate the great circle distance (in km) between two points on Earth."""
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1 
    dlon = lon2 - lon1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    r = 6371  # Radius of Earth in km
    return c * r

def min_max_scale(series):
    """Perform min-max scaling; if constant, return a series of 1's."""
    diff = series.max() - series.min()
    if diff == 0:
        return pd.Series([1] * len(series), index=series.index)
    return (series - series.min()) / diff

def weighted_cosine_similarity(u, v, weights, eps=1e-8):
    """
    Compute weighted cosine similarity between vectors u and v.
    Each element is multiplied by the square root of its weight.
    A small epsilon prevents division by zero.
    """
    weighted_u = u * np.sqrt(weights)
    weighted_v = v * np.sqrt(weights)
    dot_product = np.dot(weighted_u, weighted_v)
    norm_u = np.linalg.norm(weighted_u) + eps
    norm_v = np.linalg.norm(weighted_v) + eps
    return dot_product / (norm_u * norm_v)
def validate_user_location():
    # Assuming user_location is a global dictionary with 'lat' and 'lon' keys
    if user_location["lat"] is None or user_location["lon"] is None:
        print("⚠️ Please select a location on the map before getting recommendations.")
        return False
    return True

# We assume:
#   user_location, slider_radius, neighborhood_dropdown
#   df_coffee
#   validate_user_location, haversine_distance, min_max_scale, weighted_cosine_similarity
# are defined in previous cells.

# Radio buttons for the user's preference (hidden_gem vs hot_spot)
radio_preference = widgets.RadioButtons(
    options=['hidden_gem', 'hot_spot'],
    description='Preference:',
    style={'description_width': 'initial'}
)

# Button to trigger recommendations
button_run = widgets.Button(description="Get Recommendations", button_style='success')

# Output widget for the recommendation results
output = widgets.Output()
# Dropdown for sorting criteria
sort_options = widgets.Dropdown(
    options=['similarity_score', 'Rating', 'Total Ratings', 'Popularity Score', 'engagement', 'distance_km'],
    value='similarity_score',
    description='Sort by:',
    style={'description_width': 'initial'}
)

def run_recommendations(_):
    """Compute top coffee recommendations separately for hidden gems and hot spots."""
    with output:
        clear_output()
        
        # 1) Check if a valid location is pinned
        if not validate_user_location():
            return
        
        # 2) Extract user location & radius
        user_lat = user_location['lat']
        user_lon = user_location['lon']
        radius = slider_radius.value
        
        # 3) Copy coffee DataFrame
        df_filtered = df_coffee.copy()

        # 4) Calculate real distance
        df_filtered['distance_km'] = df_filtered.apply(
            lambda row: haversine_distance(user_lat, user_lon, row['Latitude'], row['Longitude']),
            axis=1
        )

        # 5) Filter by radius
        df_filtered = df_filtered[df_filtered['distance_km'] <= radius].copy()
        if df_filtered.empty:
            print(f"No coffee venues within {radius} km.")
            return
        
        # 6) Fill missing values
        for col in ['Rating', 'Total Ratings', 'Popularity Score', 'Total Photos', 'Total Tips']:
            df_filtered[col] = df_filtered[col].fillna(0)

        # 7) Normalize
        df_filtered['norm_rating'] = df_filtered['Rating'] / 5.0
        df_filtered['engagement'] = df_filtered['Total Photos'] + df_filtered['Total Tips']
        df_filtered['norm_engagement'] = min_max_scale(df_filtered['engagement'])
        df_filtered['norm_total_ratings'] = min_max_scale(df_filtered['Total Ratings'])
        df_filtered['norm_popularity'] = min_max_scale(df_filtered['Popularity Score'])

        # 8) Separate Hidden Gems & Hot Spots
        threshold = 0.5  # Popularity/engagement threshold

        df_hidden_gems = df_filtered[
            (df_filtered['norm_rating'] >= 0.7) &  # High rating
            (df_filtered['norm_engagement'] < threshold) &  # Low engagement
            (df_filtered['norm_popularity'] < threshold)  # Low popularity
        ].copy()

        df_hot_spots = df_filtered[
            (df_filtered['norm_rating'] >= 0.7) &  # High rating
            ((df_filtered['norm_engagement'] >= threshold) |  # High engagement
             (df_filtered['norm_popularity'] >= threshold))  # OR high popularity
        ].copy()

        # 9) Adjust Scores for Each Category
        for df_cat, preference in [(df_hidden_gems, "hidden_gem"), (df_hot_spots, "hot_spot")]:
            if preference == "hidden_gem":
                df_cat['feature_total_ratings'] = 1 - df_cat['norm_total_ratings']
                df_cat['feature_popularity'] = 1 - df_cat['norm_popularity']
                df_cat['feature_engagement'] = 1 - df_cat['norm_engagement']
            else:
                df_cat['feature_total_ratings'] = df_cat['norm_total_ratings']
                df_cat['feature_popularity'] = df_cat['norm_popularity']
                df_cat['feature_engagement'] = df_cat['norm_engagement']

            feature_cols = [
                'norm_rating', 'feature_total_ratings', 'feature_popularity', 
                'feature_engagement'
            ]
            X_features = df_cat[feature_cols].values
            weights = np.array([1.0, 1.0, 1.0, 1.0])
            ideal_vector = np.ones(len(feature_cols))

            df_cat['base_score'] = [weighted_cosine_similarity(ideal_vector, x, weights) for x in X_features]
            df_cat['base_score'] *= 0.8  # Scale base score

            df_cat['nbhd_bonus'] = np.where(
                df_cat['Nearest Neighborhood'] == neighborhood_dropdown.value[0],
                0.2,
                0.0
            )

            df_cat['similarity_score'] = (df_cat['base_score'] + df_cat['nbhd_bonus']).clip(0, 1)

            # 10) Get Sorting Preference from User
        sort_by = sort_options.value

        # 11) Determine Sort Order (Ascending for Distance, Descending for Everything Else)
        ascending_order = True if sort_by == 'distance_km' else False

        # 12) Sort Separately:
        df_hidden_gems = df_hidden_gems.sort_values(by=sort_by, ascending=ascending_order).head(10)
        df_hot_spots = df_hot_spots.sort_values(by=sort_by, ascending=ascending_order).head(10)

        # 13) Display Both Separately
        print("\n🔥 **Hot Spots (Highly Rated & Popular):**")
        display(df_hot_spots[['Name', 'Rating', 'Total Ratings', 'Popularity Score', 'engagement', 'distance_km', 'similarity_score']])

        print("\n💎 **Hidden Gems (Highly Rated but Less Popular):**")
        display(df_hidden_gems[['Name', 'Rating', 'Total Ratings', 'Popularity Score', 'engagement', 'distance_km', 'similarity_score']])


# Link the button to the function
button_run.on_click(run_recommendations)

# Show widgets & output
display(radio_preference, sort_options, button_run, output)
```

```python
import pandas as pd
import numpy as np
from math import radians, cos, sin, asin, sqrt
import ipywidgets as widgets
from IPython.display import display, clear_output

# We assume:
#   user_location, slider_radius, neighborhood_dropdown
#   df_rest, validate_user_location, haversine_distance, min_max_scale, weighted_cosine_similarity
# are defined in previous cells.

all_restaurant_categories = sorted(df_rest['Category'].dropna().unique())
category_select_rest = widgets.SelectMultiple(
    options=all_restaurant_categories,
    description='Categories:',
    style={'description_width': 'initial'}
)

radio_preference_rest = widgets.RadioButtons(
    options=['hidden_gem','hot_spot'],
    description='Preference (Restaurants):',
    style={'description_width': 'initial'}
)

button_rest = widgets.Button(
    description="Get Restaurant Recommendations", 
    button_style='success'
)
output_rest = widgets.Output()

def run_restaurant_recommendations(_):
    with output_rest:
        clear_output()

        if not validate_user_location():
            return
        
        lat_user = user_location['lat']
        lon_user = user_location['lon']
        chosen_radius_rest = slider_radius.value
        preference_rest = radio_preference_rest.value
        
        df_filtered_rest = df_rest.copy()

        chosen_categories = list(category_select_rest.value)
        if chosen_categories:
            df_filtered_rest = df_filtered_rest[
                df_filtered_rest['Category'].isin(chosen_categories)
            ].copy()

        if df_filtered_rest.empty:
            print("No restaurants found for the chosen category/categories.")
            return

        # 1) Distance
        df_filtered_rest['distance_km'] = df_filtered_rest.apply(
            lambda row: haversine_distance(lat_user, lon_user, row['Latitude'], row['Longitude']),
            axis=1
        )
        df_filtered_rest = df_filtered_rest[df_filtered_rest['distance_km'] <= chosen_radius_rest].copy()
        if df_filtered_rest.empty:
            print(f"No restaurants within {chosen_radius_rest:.1f} km.")
            return

        # 2) Distance score
        df_filtered_rest['distance_score'] = df_filtered_rest['distance_km'].apply(
            lambda d: max(0, 1 - ((d if d >= 2 else 2) / chosen_radius_rest))
        )

        # 3) Fill missing
        for col in ['Rating','Total Ratings','Popularity Score','Total Photos','Total Tips']:
            df_filtered_rest[col] = df_filtered_rest[col].fillna(0)

        # 4) Normalize rating, popularity, engagement
        df_filtered_rest['norm_rating'] = df_filtered_rest['Rating'] / 5.0
        df_filtered_rest['engagement'] = df_filtered_rest['Total Photos'] + df_filtered_rest['Total Tips']
        df_filtered_rest['norm_engagement'] = min_max_scale(df_filtered_rest['engagement'])
        df_filtered_rest['norm_total_ratings'] = min_max_scale(df_filtered_rest['Total Ratings'])
        df_filtered_rest['norm_popularity'] = min_max_scale(df_filtered_rest['Popularity Score'])

        # 5) hidden_gem vs hot_spot
        if preference_rest == "hidden_gem":
            ft_total_ratings_rest = 1 - df_filtered_rest['norm_total_ratings']
            ft_popularity_rest = 1 - df_filtered_rest['norm_popularity']
            ft_engagement_rest = 1 - df_filtered_rest['norm_engagement']
        else:
            ft_total_ratings_rest = df_filtered_rest['norm_total_ratings']
            ft_popularity_rest = df_filtered_rest['norm_popularity']
            ft_engagement_rest = df_filtered_rest['norm_engagement']

        df_filtered_rest['feature_total_ratings_rest'] = ft_total_ratings_rest
        df_filtered_rest['feature_popularity_rest'] = ft_popularity_rest
        df_filtered_rest['feature_engagement_rest'] = ft_engagement_rest

        # 6) Weighted Cosine Similarity => "base_score" up to 1
        feature_cols_rest = [
            'norm_rating',
            'feature_total_ratings_rest',
            'feature_popularity_rest',
            'feature_engagement_rest',
            'distance_score'
        ]
        X_features_rest = df_filtered_rest[feature_cols_rest].values

        weights_rest = np.array([1.0]*5)
        ideal_vec_rest = np.ones(len(feature_cols_rest))
        sim_scores_rest = [
            weighted_cosine_similarity(ideal_vec_rest, vec, weights_rest)
            for vec in X_features_rest
        ]
        df_filtered_rest['base_score_rest'] = sim_scores_rest

        # 7) Neighborhood bonus => separate
        #    We'll treat "base_score_rest" as up to 0.8, and neighborhood up to 0.2
        #    1) Scale base_score_rest so its max is 0.8
        df_filtered_rest['base_score_rest'] *= 0.8  # now max possible is 0.8

        #    2) If same neighborhood => +0.2, else +0
        if 'Nearest Neighborhood' in df_filtered_rest.columns and 'neighborhood_dropdown' in globals():
            chosen_nbhd = neighborhood_dropdown.value[0]
            df_filtered_rest['nbhd_bonus_rest'] = np.where(
                df_filtered_rest['Nearest Neighborhood'] == chosen_nbhd,
                0.2,  # the "bonus"
                0.0
            )
        else:
            df_filtered_rest['nbhd_bonus_rest'] = 0.0

        # 8) Final similarity = base + bonus
        df_filtered_rest['similarity_score_rest'] = (
            df_filtered_rest['base_score_rest'] + df_filtered_rest['nbhd_bonus_rest']
        ).clip(upper=1.0)  # never exceed 1

        # 9) Sort top 10
        global df_recommendations_rest
        df_recommendations_rest = df_filtered_rest.sort_values(
            by='similarity_score_rest', ascending=False
        ).head(10)

        display_cols = [
            'Name','Category','Rating','Total Ratings','Popularity Score',
            'distance_km','Nearest Neighborhood','similarity_score_rest'
        ]
        display_cols = [c for c in display_cols if c in df_recommendations_rest.columns]
        display(df_recommendations_rest[display_cols])

button_rest.on_click(run_restaurant_recommendations)
display(category_select_rest, radio_preference_rest, button_rest, output_rest)
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import numpy as np

# A SelectMultiple for just "Shopping Mall" and/or "Movie Theater"
category_select_venue = widgets.SelectMultiple(
    options=["Shopping Mall", "Movie Theater"],
    description='Venue Categories:',
    style={'description_width': 'initial'}
)

# Button & output for the recommendations
button_venue = widgets.Button(
    description="Get Venue Recommendations",
    button_style='success'
)
output_venue = widgets.Output()

def run_venue_recommendations(_):
    with output_venue:
        clear_output()
        
        # Ensure we have a valid pinned location
        if not validate_user_location():
            return
        
        # Grab lat/lon & radius
        lat_user = user_location['lat']
        lon_user = user_location['lon']
        radius_km = slider_radius.value
        
        # Work on a copy so we don't overwrite the original df_venue
        df_filtered_venue = df_venue.copy()
        
        # 1) Filter by chosen categories: "Shopping Mall" or "Movie Theater"
        chosen_cats = list(category_select_venue.value)
        if chosen_cats:
            df_filtered_venue = df_filtered_venue[
                df_filtered_venue['Category'].isin(chosen_cats)
            ].copy()
        
        if df_filtered_venue.empty:
            print("No venues found for the chosen categories.")
            return
        
        # 2) Compute distance to the pinned location
        df_filtered_venue['distance_km'] = df_filtered_venue.apply(
            lambda row: haversine_distance(lat_user, lon_user, row['Latitude'], row['Longitude']),
            axis=1
        )
        
        # 3) Filter by radius
        df_filtered_venue = df_filtered_venue[df_filtered_venue['distance_km'] <= radius_km].copy()
        if df_filtered_venue.empty:
            print(f"No venues found within {radius_km:.1f} km.")
            return
        
        # 4) Create a distance score in [0..1], 
        #    1 => distance=0, 0 => distance >= radius
        df_filtered_venue['distance_score'] = df_filtered_venue['distance_km'].apply(
            lambda d: max(0, 1 - (d / radius_km))
        )
        
        # 5) Identify same-neighborhood (1) vs. different (0)
        #    If we have "Nearest Neighborhood" & a global "neighborhood_dropdown"
        if 'Nearest Neighborhood' in df_filtered_venue.columns and 'neighborhood_dropdown' in globals():
            chosen_neighborhood_name = neighborhood_dropdown.value[0]
            df_filtered_venue['same_nbhd_flag'] = np.where(
                df_filtered_venue['Nearest Neighborhood'] == chosen_neighborhood_name, 
                1.0, 
                0.0
            )
        else:
            # If we don't have that info, treat all as 0
            df_filtered_venue['same_nbhd_flag'] = 0.0
        
        # 6) Combine them with a weighting system that never exceeds 1
        #    e.g. 0.7 => same neighborhood, 0.3 => distance
        w_nbhd = 0.6
        w_dist = 0.4
        df_filtered_venue['similarity_score_venue'] = (
            w_nbhd * df_filtered_venue['same_nbhd_flag'] +
            w_dist * df_filtered_venue['distance_score']
        )
        # This ensures the max possible is 0.7 + 0.3 = 1.0
        
        # 7) Sort by final score, top 10
        global df_recommendations_venue
        df_recommendations_venue = df_filtered_venue.sort_values(
            by='similarity_score_venue', ascending=False
        ).head(10)
        
        # 8) Display relevant columns
        cols_to_show = [
            'Name', 'Category', 'distance_km', 
            'Nearest Neighborhood', 'similarity_score_venue'
        ]
        cols_to_show = [c for c in cols_to_show if c in df_recommendations_venue.columns]
        
        display(df_recommendations_venue[cols_to_show])

button_venue.on_click(run_venue_recommendations)

display(category_select_venue, button_venue, output_venue)
```
