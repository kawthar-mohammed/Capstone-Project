```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup

# Initialize WebDriver
driver = webdriver.Chrome()

# Read the CSV file containing URLs
urls_activitesc = pd.read_csv('Data/urls_data/url_list_ar_festival.csv', header=None)
urls_activitesc.columns = ['url']
urls_activitesc['type'] = 'activite'

# Create an empty DataFrame to store event data
event_data = pd.DataFrame(columns=['names', 'price', 'date', 'location', 'description', 'location_url', 'type'])

# Loop through each URL
for index, row in urls_activitesc.iterrows():
    url = row['url']
    driver.get(url)

    # Wait for the page to load
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, "event-item__name-title"))
    )

    # Parse the page
    soup = BeautifulSoup(driver.page_source, "html.parser")

    # Extract name
    name_div = soup.find('h1', class_="event-item__name-title")
    name = name_div.text.strip() if name_div else 'N/A'

    # Extract price
    price_div = soup.find("div", class_="buy-block__price")
    price = price_div.text.strip() if price_div else 'N/A'

    # Extract date
    date_p = soup.find("div", class_="buy-block__date")
    date = date_p.text.strip() if date_p else 'N/A'

    # Extract description
    description_p = soup.find("div", class_="description-structured__text description-structured__text--mce description-structured__text--overview")
    description = description_p.decode_contents() if description_p else 'N/A'

    # Extract location text
    location_div = soup.find("a", class_="location-block__venue-name--link")
    location_text = location_div.text.strip() if location_div else 'N/A'

    # Extract location URL
    google_maps_link = soup.find("a", class_="location-block__view-navigation-button link link--blue")
    location_url = google_maps_link['href'] if google_maps_link else 'N/A'

    # Store in DataFrame
    new_row = pd.DataFrame({
        'names': [name],
        'price': [price],
        'date': [date],
        'location': [location_text],
        'description': [description],
        'location_url': [location_url],
        'type': ['activite']
    })

    event2_data = pd.concat([event2_data, new_row], ignore_index=True)

# Save the data
event2_data.to_csv("event2_data.csv", index=False)

# Close the WebDriver
driver.quit()
```
