# Get Urls links form webook

```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException

# Initialize the WebDriver
driver = webdriver.Chrome()

# Base URL
url = "https://webook.com/ar/explore?city=RUH&country=SA&category=activities-adventures"

# Open the webpage
driver.get(url)

# here

# Notify the user
print("The page has loaded. You can now scroll and click the 'Load More' button manually if needed.")

# Pause execution until the user is ready
input("Press Enter to continue with scraping...")

# Extract the page source after manual interaction
html = driver.page_source

# Initialize an empty list to store the URLs
url_list_activites = []

# Function to extract URLs
def extract_urls():
    try:
        WebDriverWait(driver, 30).until(
            EC.visibility_of_all_elements_located((By.XPATH, "//div[contains(@class, 'mx-auto w-full')]//a[@href]"))
        )
    except TimeoutException:
        print("Timeout: The elements did not appear in time!")
        return

    elements = driver.find_elements(By.XPATH, "//div[contains(@class, 'mx-auto w-full')]//a[@href]")

    for element in elements:
        product_url = element.get_attribute("href")
        if product_url and product_url not in url_list_activites:
            url_list_activites.append(product_url)

# Extract URLs from the current page
extract_urls()

# Print the collected URLs
print("Collected URLs:", url_list_activites)

# Close the WebDriver
driver.quit()
```

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup

# Initialize the WebDriver
driver = webdriver.Chrome()

# Read the CSV file containing URLs (assuming the CSV file has one column with URLs)
urls_activitesc = pd.read_csv('url_data/url_list_activites.csv', header=None)  # No header for one-column CSV

# Rename the column for easier access
urls_activitesc.columns = ['url']

# Add a new column 'type' with the value 'activite' for all rows
urls_activitesc['type'] = 'activite'


# Loop through each URL in the URLs from the CSV file
for index, row in urls_activitesc.iterrows():
    url = row['url']  # Access the URL

    # Request the page
    driver.get(url)

    # Wait for the page to load (you can adjust the sleep time if needed)
    time.sleep(3)

    # Parse the page with BeautifulSoup
    soup = BeautifulSoup(driver.page_source, "html.parser")

    # Extract product names from the page using the specific class for the title
    name_div = soup.find('h1', class_='text-heading-L font-heading leading-tight text-text-primary')

    # Extract price
    price_div = soup.find("div", class_="group flex items-center gap-2")
    price = price_div.find("span").text if price_div else 'N/A'  # Extract price from span tag

    # Extract date, location, and description
    date_p = soup.find("p", class_="text-body-S font-body text-text-secondary")
    description_p = soup.find("p", class_="my-1")

    # Extract the second location from the divs with class `flex items-center gap-1.5 text-sm grow`
    location_divs = soup.find_all("div", class_="flex items-center gap-1.5 text-sm grow")
    location_text = 'N/A'
    location_url = 'N/A'

    # Ensure there are at least two location divs
    if len(location_divs) > 1:
        location_a = location_divs[1].find("a")  # Get the second div (index 1)
        if location_a:
            location_url = location_a['href']  # Extract the URL from the 'href' attribute
            location_text = location_a.find("p").text.strip()  # Extract the location name from the <p> tag

    # Extracting the full description including any <a> tags
    description = 'N/A'
    if description_p:
        description = description_p.decode_contents()  # Get the HTML content including <a> tags

    # Extract text for each field, or set to 'N/A' if the tag is not found
    date = date_p.text.strip() if date_p else 'N/A'
    name = name_div.text.strip() if name_div else 'N/A'

    # Store the extracted data in a DataFrame
    new_row = pd.DataFrame({
        'names': [name],
        'price': [price],
        'date': [date],
        'location': [location_text],
        'description': [description],
        'location_url': [location_url],
        'type': ['activite']  # Add the type column with value 'activite'
    })

    # Concatenate with the main DataFrame
    event_data = pd.concat([event_data, new_row], ignore_index=True)


# Close the WebDriver
driver.quit()
```

```python
import pandas as pd
from urllib.parse import urlparse, parse_qs

# Function to extract latitude and longitude from a Google Maps URL
def extract_lat_lng(google_maps_url):
    # Parse the URL
    parsed_url = urlparse(google_maps_url)
    
    # Extract the query parameters
    query_params = parse_qs(parsed_url.query)
    
    # Extract latitude and longitude from the query parameter 'query'
    if 'query' in query_params:
        # The query parameter should be in the format "lat,lng"
        lat_lng = query_params['query'][0]  # Get the first item in the list
        lat, lng = lat_lng.split(',')
        return float(lat), float(lng)
    else:
        return None, None



# Apply the function to extract Latitude and Longitude
event_data[['Latitude', 'Longitude']] = event_data['location_url'].apply(
    lambda url: pd.Series(extract_lat_lng(url))
)

event_data
```
